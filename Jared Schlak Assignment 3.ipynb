Problem 1
In Unix, you can execute a command in two ways:
•Entering its name, e.g., “cp” or “ls”.
•Entering “! <index>”. This notation is used to repeat the indexth (1-based) command since
the start of the session. For example, suppose that the user has entered the following
commands:
ls, cp, mv, mv, mv, !1, !3, !6.
“!1” would trigger the execution of “ls”, “!3” would repeat “mv” and “!6” would execute “!1”
which in turn would trigger the execution of “ls”.
You are provided with a sequence of commands that the user has entered in the terminal
since the start of the session. Each command can be one of the following: “cp”, “ls”, “mv” or
“! <index>”. Calculate the number of times each of “cp”, “ls” and “mv” commands are executed
and return the array of three integers in the following form: [# of times for “cp”, # of times for “ls”,
# of times for “mv”].

```python
array = [ "ls", "cp", "mv", "mv", "mv", "!1", "!3", "!6"]

def commands_count (List):
    list = [0, 0, 0]
    for i in range(len(List)):
        if "!" in List[i]:
                t = int((List[i])[1:])-1
                List[i] = List[t]
        if List[i] == "cp":
            list[0] += 1
        elif List[i] == "ls":
            list[1] += 1
        elif List[i] == "mv":
            list[2] += 1
    print(List)
    return list

commands_count(array)

```

Problem 4
A basket has ‘n’ fruits of two types - apples and oranges. They decide to play a game. The
game is played in multiple rounds until one of the fruit types wins (either apple or orange). In
each round, every ‘eligible’ fruit gets to perform one of the two actions.
Action1:
Disqualify one member of another fruit type from the game, so that they cannot perform any
action in the current round or any subsequent rounds. (Disqualified fruit becomes ‘ineligible’
and cannot perform any actions)
Action2:
If all members of another fruit type are disqualified(ineligible), declare his fruit type as the winner
of the game.
Every round begins with fruits performing actions in the given order, and every fruit chooses
the action best for his fruit type, predict which fruit type will be the winner!. Time complexity
O(n) and space complexity O(n)
Input: A string s of length ‘n’ representing order in which ‘n’ fruits can perform an action in
every round. (‘A’: apple, ‘O’: orange) Output: Winner type: string (either “apple” or “orange”)

```python
def predict_winner(s):
    #  ans = "apple" or "orange"
    appleindexarray = []
    orangeindexarray = []
    for i in range(len(s)):
        if s[i] == "A":
            appleindexarray.append(i)
        else:
            orangeindexarray.append(i)
    print(appleindexarray, orangeindexarray)
    count = 0
    nextcount = len(s)
    while appleindexarray != [] and orangeindexarray != []:
        if count in appleindexarray:
            orangeindexarray.pop(0)
            appleindexarray.pop(0)
            appleindexarray.append(nextcount)
            nextcount += 1
            print(appleindexarray, orangeindexarray)
        elif count in orangeindexarray:
            appleindexarray.pop(0)
            orangeindexarray.pop(0)
            orangeindexarray.append(nextcount)
            nextcount += 1
            print(appleindexarray, orangeindexarray)
        count += 1
    if appleindexarray == []:
        winner = "orange"
    elif orangeindexarray == []:
        winner = "apple"
    return winner
                

s = "AAAOOOOAOOAAAAO"
predict_winner(s)
```
